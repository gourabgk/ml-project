# Create Training and Testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

# Import Different Regression models

```python
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import BayesianRidge # Bayesian algorithm
from sklearn.linear_model import ElasticNet # ElasticNet algorithm
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

# Random Forest Regressor

```python
rf_random=RandomForestRegressor(n_estimators=100,random_state=0)
rf_random.fit(x_train,y_train)
```

```python
y_predict_rfr=rf_random.predict((x_test))
print(y_predict_rfr)
r_squre=r2_score(y_test,y_predict_rfr)
print("R squre error asssociated with random forest regressor is",r_squre)
```

# Graphical Representation of Accuracy

```python
plt.scatter(y_test,y_predict_rfr,color='coral')
```

```python
import pickle
#open a file,where you ant to store the data
file=open("random.pkl","wb")

#dump information to that file
pickle.dump(rf_random,file)
file.close()
```

# Linear Regression

```python
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(x_train, y_train)  # perform linear regression
Y_pred = linear_regressor.predict((x_test))
print(Y_pred)
```

```python
r_squreS=r2_score(y_test,Y_pred)
print("R squre error asssociated with Linear regressor is",r_squreS)
```

```python
file1=open("linears.pkl","wb")

#dump information to that file
pickle.dump(linear_regressor,file1)
file1.close()
```

# Lasso Regression

```python
model_lasso = Lasso(alpha=0.01)
model_lasso.fit(x_train, y_train) 
pred_test_lasso= model_lasso.predict((x_test))
print(pred_test_lasso)
```

```python
print("R squre error asssociated with Lasso regressor is",r2_score(y_test, pred_test_lasso))
```

```python
file2=open("lassos.pkl","wb")
#dump information to that file
pickle.dump(model_lasso,file2)
file2.close()
```

```python
rr = Ridge(alpha=0.01)
rr.fit(x_train, y_train)
pred_test_rr= rr.predict(x_test)
print(pred_test_rr)
print("R squre error asssociated with Ridge regressor is",r2_score(y_test, pred_test_rr))
```

```python
file3=open("ridges.pkl","wb")
#dump information to that file
pickle.dump(rr,file3)
file3.close()
```

```python


# 4. Bayesian

bayesian = BayesianRidge()
bayesian.fit(x_train, y_train)
bayesian_yhat = bayesian.predict(x_test)
print(bayesian_yhat)

print("R squre error asssociated with Bayesian Ridge regressor is",r2_score(y_test, bayesian_yhat))


```

```python
file4=open("bayesian.pkl","wb")
#dump information to that file
pickle.dump(bayesian,file4)
file4.close()
```

```python

# 5. ElasticNet
en = ElasticNet(alpha = 0.01)
en.fit(x_train, y_train)
en_yhat = en.predict(x_test)
print(en_yhat)
print("R squre error asssociated with Elastic Net regressor is",r2_score(y_test, en_yhat))
```

```python
file5=open("Elastic.pkl","wb")
#dump information to that file
pickle.dump(en,file5)
file5.close()
```
